```python
data = '../input/mydata/data1.txt' #Getting the location of the data
```

```python
list_native_bal = [] #Creating an empty list for storing native balances
list_non_native_bal = [] #Creating an empty list for storing non-native balances

with open(data) as file1:
    string = file1.read() #Reading the content of the .txt file
    list1 = string.split(',')
    for i in list1:
        if 'nativeBalance' in i:
            native_balance = i[(len('nativeBalance') + 4):(len(i) - 1)]
            list_native_bal.append(float(native_balance))
        elif 'nonNativeBalance' in i:
            non_native_balance = i[(len('nonNativeBalance') + 4):(len(i) - 1)]
            list_non_native_bal.append(float(non_native_balance))
        else:
            pass
        
print("Non Native Balances:")

#Printing 10 samples each for native and non native balances
for i in range(10):
    print(list_non_native_bal[i])

print("Native Balances: ")

for i in range(10):
    print(list_native_bal[i])
```

```python
#Importing the modules required for plotting the graphs
import numpy as np 
import matplotlib.pyplot as plt
```

```python
non_native_np = np.array(list_non_native_bal) #Converting list to numpy array
#Checking if all is as before
for i in range(10):
    print(non_native_np[i])
```

```python
native_np = np.array(list_native_bal) #Converting list to numpy array
#Checking if all is as before
for i in range(10):
    print(native_np[i])
```

```python
#Now let's plot some beautiful graphs on the basis of the data collected above
fig = plt.figure()
plt.title("A 2-D Graph between native and non-native balances")
plt.xlabel("Native Balance")
plt.ylabel("Non-Native Balance")
plt.scatter(native_np, non_native_np)
plt.show()
```

```python
plt.title("2-D Bar Graph of native balances")
plt.hist(native_np, height = 15000)
```

```python
plt.title("2-D Bar Graph of non-native balances")
plt.hist(non_native_np, height = 15000)
```
